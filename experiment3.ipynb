## Executions 

The ABM has been explored over two parameters ($\beta$ and $\eta$), after fixing three of them to the most stable values (*g*, *k* and *w*):
The parameters appear in the ABM as:

- *k* is the capital intensity, fixed to 1.0.
- *w* is the real wage, fixed to 0.7.
- *g* is the bankruptcy cost, fixed to 1.2.
- $\beta$ is the skewness, varying in [ 0.02, 0.03, 0.04, 0.05].
- $\eta$ is the inverse of the elasticity of demand ($1/\varepsilon$), varying in [0.0001, 0.1, 0.3, 0.5, 0.8].

For each combination of the varying parameters, the model has been run 10 times using a Montecarlo method. So a total of 200 executions of the model have been done (20 combinations, 10 times each).

## Importance of $\beta$
Till now, we have vary eta having fixed $\beta$ to 0.02. Let's repeat the previous work, but fixing eta to 0.0001 and discover how affects $\beta$ to our model. But if we look at GDP, the impact is not so clear, as @fig-gdp-eta00gdp shows:

![Fixing eta=0.0001, and exploring GDP varying beta=0.02 (top left), beta=0.03 (top right), eta=0.04 (bottom left) and eta=0.05 (bottom right)](experiment3b\beta_avgy.pdf){#fig-gdp-eta00gdp}

![Same as previous figure, but with profits.](experiment3b\beta_avgprof.pdf){#fig-gdp-eta00profits}

$\beta$ is also relevant, because is connected with the determination of the interest rate, as @fig-gdp-eta00r shows: greater the $\beta$, greater the average $r$, and logically it should generate greater $\gamma$, which generates less $K$ and so on.

```python
#| label: fig-gdp-eta00r
#| fig-cap: "Fixing eta=0.0001, and exploring average interest rates varying beta=0.02 (top left), beta=0.03 (top right), eta=0.04 (bottom left) and eta=0.05 (bottom right)"
#| echo: false
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 15), sharex=True, sharey=True)
ax1.imshow(mpimg.imread('experiment3b\\r_beta002.png'))
ax1.axis('off')
ax2.imshow(mpimg.imread('experiment3b\\r_beta003.png'))
ax2.axis('off')
ax3.imshow(mpimg.imread('experiment3b\\r_beta004.png'))
ax3.axis('off')
ax4.imshow(mpimg.imread('experiment3b\\r_beta005.png'))
ax4.axis('off')
```

From the point of view of the bank sector, the impact of 𝛽 may could be present in the profits:

$$ 
\pi_{b,t}=\sum{r_{i,t}L_{i,t}-r(D_t+A_{b,t})}
$$

The profits for the bank sector have two different components: at the left of the addition the profits due to the interest rate charged to the loans to the companies. At the right, the payments the bank gives for deposits and net worth, using the average $r$ in each instant of time. 
In @fig-betabank we can realize that this two components have similar response after altering $\beta$. 

![Fixing eta=0.0001, and exploring beta: profits of the bank sector are produced only when blue balloons, when difference between green line (revenues of loans) and red line (payments of deposits and net worth) are positive, not always possible](Experiment3b\beta.pdf){#fig-betabank} 


At the end, from the point of view of the firms, one of the effects is the increment in bankruptcies
in the firms when we increment 𝛽, as @fig-failures-beta shows:


```python
#| label: fig-failures-beta
#| fig-cap: "Average number of failures varying beta, for eta=0.0001 (top left), eta=0.1 (top right), eta=0.3 (middle left), eta=0.5 (middle right) and eta=.08 (bottom left)"
#| echo: false
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(20, 20), sharex=True, sharey=True)
ax1.imshow(mpimg.imread('experiment3b\\eta0.0001_FirmsFAIL.png'))
ax1.axis('off')
ax2.imshow(mpimg.imread('experiment3b\\eta0.1_FirmsFAIL.png'))
ax2.axis('off')
ax3.imshow(mpimg.imread('experiment3b\\eta0.3_FirmsFAIL.png'))
ax3.axis('off')
ax4.imshow(mpimg.imread('experiment3b\\eta0.5_FirmsFAIL.png'))
ax4.axis('off')
ax5.imshow(mpimg.imread('experiment3b\\eta0.8_FirmsFAIL.png'))
ax5.axis('off')
ax6.axis('off')
```

The impact is not comparable to the number of failures varying eta (@fig-etabeta) but needs an explanation: the increment in interest rate driven by beta provokes a reduction (not really important, but present) in the benefits of the bank sector, that drives to a reduction in net worth that generates in next iterations to minorate the loans and so, problems for the firms.

```python
#| label: fig-etabeta
#| fig-cap: "Average number of failures varying beta for eta=0.0001 (left) and varying eta for beta=0.02 (right)"
#| echo: false
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
fig, ((ax1, ax2)) = plt.subplots(1, 2, sharex=True, figsize=(15, 15), sharey=True)
ax1.imshow(mpimg.imread('experiment3b\\eta0.0001_FirmsFAIL.png'))
ax1.axis('off')
ax2.imshow(mpimg.imread('experiment3a\\beta0.02_FirmsFAIL.png'))
ax2.axis('off')
```
